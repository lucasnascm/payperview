```python
for b in brothers:
  L = Instaloader()
  profile = Profile.from_username(L.context, b)
  periodo = (datetime.now() - timedelta(hours = 3)).strftime('%Y-%m-%d %H:%M:%S')
  seg = profile.followers
  seguidores_bbb = seguidores_bbb.append({'periodo':periodo, 'perfil':b, 'seguidores':seg}, ignore_index=True)

seguidores_bbb.to_csv('/content/seguidores_bbb.csv', na_rep='', index=False)
#exec("seguidores_bbb.to_csv('/content/seguidores_bbb_{}.csv', na_rep='', index=False)".format(periodo))
seguidores_bbb
```
